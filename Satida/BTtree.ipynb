```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Dec 18 02:29:29 2019

@author: satida
"""
import sys 
sys.setrecursionlimit(50000)

class BTree:
    """
    Satida.Btree.BTree 
    --------------------
    
    Params :
    ----------------------
    
    mode = {Simple : the easy way and insert every word into Btree
    
        \t  Normal : String of complete Dictionary and insert every delimiter index.
        
        \t  Full : all dictionary in blocks with k size. insert references index.
        
        \t  Extra : all dictionary in blocks with k size and at first prefix chars and insert references index into btree }
    """
    def __init__(self,mode):
        modes={"Simple":self.SimpleBtree ,"Normal":self.NormalBtree,"Full":self.FullBtree,"Extra":self.ExtraBtree}
        try :
            self.mode=modes[mode]
            self.name=mode
        except:
            raise Exception('Mode must be in [Simple - Normal - Full - Extra ]! \n \" out of range Error.\"')


    def SimpleBtree(self):
        from .BtreeSource.SimpleBtree.Btree_Simple import Btree
        self.tree=Btree()
        self.root=self.tree.root
        self.__insert=self.tree.insert
        self.__search=self.tree.Search
        self.__returnAllValues=self.tree.ReturnAllValues
        
        
        
    def NormalBtree(self,String):
        from .BtreeSource.NormalTree.NormalCompressBtree import Btree,Converter
        self.conv=Converter()
        self.tree=Btree(String,self.conv)
        self.root=self.tree.root
        self.__insert=self.tree.insert
        self.__search=self.tree.Search
        self.__returnAllValues=self.tree.ReturnAllValues
        

        
    def FullBtree(self,String):
        from .BtreeSource.FullTree.FullCompressBtree import Btree,Converter
        self.conv=Converter()
        self.tree=Btree(String,self.conv)
        self.root=self.tree.root
        self.__insert=self.tree.insert
        self.__search=self.tree.Search
        self.__returnAllValues=self.tree.ReturnAllValues
        
        
    def ExtraBtree(self,String):
        from .BtreeSource.ExtraTree.ExtraCompressBtree import Btree,Converter
        self.conv=Converter()
        self.tree=Btree(String,self.conv)
        self.root=self.tree.root
        self.__insert=self.tree.insert
        self.__search=self.tree.Search
        self.__returnAllValues=self.tree.ReturnAllValues
        
        
    def execute(self,String):
        if self.mode != self.SimpleBtree:
            self.mode(String)
        else :
            self.mode()
        



    def Insert(self,items):
        items=self.conv.ToNorm(items)
        for i in items:
            self.__insert(i)
    
    def Search(self,item):
        res_word=list()
        res_ref=list()
        node,content=self.__search(self.tree.root,item)
        self.__returnAllValues(node,res_word,res_ref)
        return res_word,res_ref
    
    def PrintSize(self):
        import pickle,os
        save_dir="./"+self.name+".pkl"
        with open (save_dir,"wb") as file:
            pickle.dump(self.tree,file)
        statinfo = os.stat(save_dir)
        print(self.name,statinfo.st_size,"B")
            
        


















```
