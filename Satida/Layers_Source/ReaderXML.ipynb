```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Dec 10 23:51:35 2019

@author: satida
"""
import xml.etree.ElementTree as ET
import persian
from Utils.Utils import alphabet

class ReaderXML:
    
    def __init__(self,xml_dir:list,valid_tags:list):
        
        if type(xml_dir)!=list or type(valid_tags)!=list :
            raise Exception('arg2 must be a list! \" Type Checking Error.\"')
        self.xml_dirs=xml_dir
        self.valid_tags=valid_tags
        
        
    def execute(self):
        
        docs=list()
        org_Docs=list()
        for _dir in self.xml_dirs:
            xml = ET.parse(_dir)
            org_Docs += xml.findall("DOC")
            for xml_doc in xml.findall("DOC"):
                doc=""
                for tag in self.valid_tags:
                    doc += xml_doc.find(tag).text
                
                doc=persian.convert_ar_characters(persian.convert_en_numbers(doc.replace("\xa0"," ")))
                docs.append(doc)
                
        return docs,org_Docs
```
