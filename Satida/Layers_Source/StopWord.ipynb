```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Dec 11 02:10:56 2019

@author: satida
"""


class StopWord:
    
    def __init__(self,stop_words:list):
        
        if type(stop_words)!=list or type(stop_words[0])!=str:
            raise Exception('args must be a list of str! \" Type Checking Error.\"')
            
        self.stop_words=stop_words
    
    def execute(self,tokens:list):
        temp=" ".join(tokens)
        for sw in self.stop_words:
            if sw in temp:
                temp=temp.replace(" "+sw+" "," ")
        
        return temp.split(" ") 
        
        






















```
