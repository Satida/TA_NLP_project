```
# -*- coding: utf-8 -*-
"""
Created on Tue Dec 10 23:51:35 2019

@author: satida
"""

import persian

class Posting:
    
    def execute(self,main_dict,rep_dict,docs):
 
        import numpy as np
        rep_to_dict=np.zeros((len(rep_dict)),dtype=int)
        
        for i,word in enumerate(rep_dict):
            rep_to_dict[i]=main_dict.index(deleteDoller(word))
        
        posting=self.__FindPosting(docs,main_dict)
        return rep_to_dict,posting
    
    
    def __FindPosting(self,docs,main_dict):
        posting=list()
        for i in range(len(main_dict)):
            posting.append(list())
            
        for d,doc in enumerate(docs):
            for i,typ in enumerate(main_dict):
                if typ in doc:
                    posting[i].append(d)
            if int(d / len(docs) * 100 ) % 10 ==0:
                print("Posting : ",int(d / len(docs) * 100 ),"%")
                
        return posting
            
        
def deleteDoller(word:str):
    idx=word.index("$")
    return word[idx+1:]+word[:idx]        
        
    








```
