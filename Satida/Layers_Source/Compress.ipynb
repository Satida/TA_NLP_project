```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Dec 13 12:23:52 2019

@author: satida
"""

from Utils.Utils import SortFa

class Compress :
    def __init__ (self,mode:str,K = 4):
        modes={"Normal":self.NormalCompression,"Full":self.FullCompression , "Extra":self.ExtraCompression}
        try :
            self.mode=modes[mode]
            self.K=K
        except :
            raise Exception('Mode must be in [Normal - Full - Extra ]! \n \" out of range Error.\"')
    
    
    def execute(self,_dict):
        return self.mode(_dict,self.K)
    
    def NormalCompression(self,_dict,K=4):
        Sdict=SortFa(_dict)
        indexes=list()
        String=",".join(Sdict)
        for i,ch in enumerate(String):
            if i == 0 or ch ==",":
                indexes.append(i)
                
        return {"String":String,"indexes":indexes}
    
    def FullCompression (self,_dict,K=4):
        String=[]
        pointer=[]
        j=0
        l=0
        for i,word in enumerate(_dict):
            if i % K == 0 and i != 0:
                String.append(",".join(_dict[j:i]))
                pointer.append(l)
                l += 1
                j = i
                
        return {"String":String,"indexes":pointer}
    
    
    
    
    def ExtraCompression (self,_dict,K=4):
        result=self.FullCompression(_dict)
        block_list=result["String"]
        newStringBlock=[]
        for block in block_list:
            prefix=""
            
            for i in range(min([len(j) for j in block.split(",")])):
                char = block.split(",")[0][i]
                uniq=set()
                
                for word in block.split(","):
                    uniq.add(word[i])
                if len(uniq) == 1:
                    prefix += char
                else :
                    break 
            res=prefix+"%"+",".join([word.replace(prefix,"") for word in block.split(",")])
            
            newStringBlock.append(res)
                
                    
        return {"String":newStringBlock,"indexes":result["indexes"]}
            
            
            











```
